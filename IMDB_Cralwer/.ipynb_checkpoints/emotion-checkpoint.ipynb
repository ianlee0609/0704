```python
import requests
from bs4 import BeautifulSoup as bs
import pyodbc

conn = pyodbc.connect('DRIVER={SQL Server};SERVER=163.14.72.42;DATABASE=FB_IMDB;UID=sa;PWD=scucc')
cursor = conn.cursor()


res= requests.get('https://en.wikipedia.org/wiki/List_of_emoticons')

soup = bs(res.text)

ALLtable= soup.select('.wikitable')
for page in range(0,3):
    if page==0:
        kind='Western'
    if page==1:
        kind='Eastern'
    if page==2:
        kind='2channel emoticons'
    table=ALLtable[page]
    for title in table.findAll('tr'):
        i=1
        for message in title.findAll('td'):
            content=message.text
            if i==1:
                 emotion= content.split(' ')

            if i==2:
                while content.find('[')!=-1:

                    initial=content.find('[')
                    end= content.find(']')

                    content= content[:initial]+content[end+1:]
                meaning= content
            i+=1
        
        for emo in emotion:
            print emo,meaning,kind
            insertSQL()
            conn.commit()
        
        
        
       
cursor.close()
conn.close()  
```
