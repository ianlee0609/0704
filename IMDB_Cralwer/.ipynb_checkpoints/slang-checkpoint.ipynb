```python
import requests
from bs4 import BeautifulSoup as bs
res = requests.get('http://slangit.com/terms/social_media')


slang=""
meaning=""
popularity=""

soup = bs(res.text)
for title in soup.select('tr'):
    i=1
    for a in title.select('td'):
        if i==1:
            slang=a.text
        if i==2:
            meaning=a.text
        if i==3:
            popularity= str(a).count('star full')+str(a).count('star half')*0.5
            break
        i+=1
        
    print slang+"\n"+meaning
    print  popularity
```
