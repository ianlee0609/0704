```python
def getPicture(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    for url2 in soup.select('img'):
        if 'jpg' in url2['src'] and 'cartoonmad' in url2['src']:
            res2 = requests.get(url2['src'], stream=True)
            with open('comic/{}'.format(url2['src'].split('/')[-1]), 'wb') as f:
                shutil.copyfileobj(res2.raw, f)
```

```python
from bs4 import BeautifulSoup as bs
dic = {'機關代碼': '', '機關名稱': '', '單位名稱': '' , '機關地址': '', '聯絡電話': ''}
with open('gov/51787884.txt','r') as f:
    soup = bs(f.read())
    for tr in soup.select('tr'):
        th = tr.select('th')
        if len(th) > 0 :
            th_value = th[0].text.encode('utf-8').strip()
            if th_value in dic:
                td = tr.select('td')[0].text.strip()
                dic[th_value] = td
```

```python
import requests
import jieba
dic = {}
for page in xrange(1,5):
    res = requests.get('http://ent.appledaily.com.tw/realtimenews/index/page/{}'.format(page))
    soup = bs(res.text)
    for title in soup.select('a h1'):
        for ele in jieba.cut(title.text):  
            if not ele in dic:
                dic[ele] = 1
            else:
                dic[ele] +=1 
```

```python
from operator import itemgetter
swd = sorted(dic.items(), key=itemgetter(1), reverse=True)
print swd
```
