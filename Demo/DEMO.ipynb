```python
import requests
import jieba
from collections import Counter
from bs4 import BeautifulSoup as bs

dic=Counter()
for page in range(1,3):
    res =requests.get('http://www.appledaily.com.tw/realtimenews/section/new/{}'.format(page))
    soup=bs(res.text)

    for ele in soup.select('.rtddt'):

        title=ele.select('h1')[0].text
#         print title
        for cute in jieba.cut(title):
            
                dic[cute] +=1 
for word in dic.most_common(100):
    if(len(word[0])>=2):
        print word[0],word[1]
    
```
