```python
#匯入
import requests
from bs4 import BeautifulSoup as bs
from collections import Counter
dic=Counter()

res=requests.get('http://www.chinapost.com.tw/taiwan/Default.asp?pageto=1')
soup=bs(res.text)

#頭條新聞
print '頭條新聞:' 
print soup.select('.HeadLineNewsHeader')[0].text
print '其他新聞:'


#抓1~9頁
for page in range (1,9):
    
    #讀取網頁+頁碼
    res=requests.get('http://www.chinapost.com.tw/taiwan/Default.asp?pageto='+str(page))
    #將文字檔放入BeautifulSoup中
    soup=bs( res.text)
    #找出summary_headline的class
    for ele in  soup.select('.summary_headline'):
        #我們要的title
        title= ele.text
        #空白切開
        for cut in title.split(" "):
            #如果有出現 次數+1
            dic[cut]+=1
#如果3個單字以上則印出來
for word in dic.most_common(100):
    if(len(word[0])>=3):
        print word[0],word[1]
```
